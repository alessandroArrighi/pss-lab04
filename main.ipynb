```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Paese di cittadinanza: Indica se le persone emigrate prese in considerazione sono di cittadinanza Italiana oppure non Italiana (Foreign country)
Sesso: Indica le persone di sesso maschile, femminile ed il totale (maschi + femmine)
Età: Indica l'età suddivise in gruppi: fino a 17 anni, dai 18 a 39 anni, dai 40 a 64 anni, sopra i 65 anni ed infine il totale
Value: Indica il numero di emigranti per quella relativa tupla

I continenti vanno ad indicare il numero totale di emigrazioni per tutti gli stati appartenenti a quel continente

```python
filename = "./dataframe.csv"

df = pd.read_csv(filename, sep = ',')
df.head()
```

```python
#df.to_csv('./dataframe.csv', sep = ',', index = False)
```

Quanti emigrati ci sono stati per ogni anno? Quanti fra questi hanno un età compresa tra 18 e 39 anni?

```python
# Numero di emigrati per ogni anno
tot = df.groupby(by = "Anno").sum('Numero di emigrati')
```

```python
# Numero di emigrati con età compresa tra 18 e 39 anni
età_compresa = df[df['Età'] == '18-39 anni'].groupby(by = 'Anno').sum('Numero di emigrati')
display(età_compresa)
(età_compresa / tot) * 100
```

In che percentuali si suddividono gli emigrati di cittadinanza italiana rispetto agli emigrati di cittadinanza straniera? Quali sono le percentuali per i soggi di età compresa tra 18 e 39 anni?

```python
# Percentuale per ogni anno degli emigrati di cittadinanza italiana e di cittadinanza straniera

ita = (df[df['Paese di cittadinanza'] == 'Italia'].groupby(by = "Anno").sum('Numero di emigrati') / tot) * 100
display(ita)
foreign = (df[df['Paese di cittadinanza'] == 'Foreign country'].groupby(by = "Anno").sum('Numero di emigrati') / tot) * 100
foreign
```

```python
# Percentuali per i soggetti di età compresa tra 18 e 39 anni

ita = df[df['Paese di cittadinanza'] == 'Italia'].groupby(by = "Anno").sum('Numero di emigrati')
display((df.loc[(df['Età'] == '18-39 anni') & (df['Paese di cittadinanza'] == 'Italia')].groupby(by = "Anno").sum('Numero di emigrati') / ita) * 100)
(df.loc[(df['Età'] == '18-39 anni') & (df['Paese di cittadinanza'] == 'Italia')].groupby(by = "Anno").sum('Numero di emigrati') / tot) * 100
```
